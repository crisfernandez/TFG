```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Oversampling

```python
# Participantes
participants = ['P3', 'P6', 'P14', 'P19']

# Valores de F1-Score
f1_base = [0.480, 0.540, 0.520, 0.450]
f1_random = [0.550, 0.580, 0.570, 0.650]
f1_smote = [0.620, 0.670, 0.590, 0.610]

# Valores ROC-AUC
roc_base = [0.558, 0.625, 0.652, 0.571]
roc_random = [0.603, 0.656, 0.749, 0.761]
roc_smote = [0.685, 0.760, 0.775, 0.709]

x = np.arange(len(participants))
width = 0.2


```

```python
x = np.arange(len(participants))
width = 0.25

palette = sns.color_palette("Pastel1", 3)
# F1-score plot
plt.figure(figsize=(8,4))
plt.bar(x - width, f1_base, width, label='AttentionBaseNet base', color=palette[0])
plt.bar(x, f1_random, width, label='Aleatorio', color=palette[1])
plt.bar(x + width, f1_smote, width, label='SMOTE', color=palette[2])
plt.xticks(x, participants)
plt.ylabel('F1-score')
plt.title('Comparaci√≥n de F1-score')
plt.legend(loc='upper left', frameon=False)

plt.tight_layout()
plt.show()
```

```python
# ROC-AUC 
plt.figure(figsize=(8,4))
plt.bar(x - width, roc_base, width, label='AttentionBaseNet base', color=palette[0])
plt.bar(x, roc_random, width, label='Aleatorio', color=palette[1])
plt.bar(x + width, roc_smote, width, label='SMOTE', color=palette[2])
plt.xticks(x, participants)
plt.ylabel('ROC-AUC')
plt.title('Comparaci√≥n de ROC-AUC')
plt.legend(frameon=False)
plt.tight_layout()
plt.show()
```
