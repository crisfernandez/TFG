```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Desbalanceados

```python
data = {
    'Participante': ['P3', 'P3', 'P6', 'P6', 'P14', 'P14', 'P19', 'P19'],
    'Clase': ['high_trust (1)', 'low_trust (0)'] * 4,
    'Muestras': [46944, 16351, 45693, 19793, 47179, 21134, 62140, 9798]
}
df_consolidado = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))
# Gráfico de barras agrupadas
sns.barplot(x='Participante', y='Muestras', hue='Clase', data=df_consolidado, palette='coolwarm')
plt.title('Gráfico 1: Balanceo de Clases')
plt.ylabel('Número de Muestras')
plt.xlabel('Participante')
plt.legend(title='Clase', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
