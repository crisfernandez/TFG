# Gráficos de la evolución del f1-score en la fase de entrenamiento de los modelos base

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('historial_f1_global.csv')  

# Participantes a visualizar
participants_to_plot = [ 'P3', 'P6', 'P19']
df_plot = df[df['participant'].isin(participants_to_plot)]

sns.set(style="whitegrid", palette="Set2")
plt.rcParams.update({'font.size': 12})

# Gráfico por arquitectura
architectures = df_plot['architecture'].unique()
for arch in architectures:
    plt.figure(figsize=(10,6))
    df_arch = df_plot[df_plot['architecture'] == arch]
    
    # Línea por participante
    sns.lineplot(data=df_arch, x='epoch', y='f1', hue='participant', marker='o')
    
    plt.title(f'Evolución del F1-score - {arch}', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('F1-score')
    plt.ylim(0,1)
    plt.xticks(df_arch['epoch'].unique())  
    plt.legend(title='Participante')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show() 
```
